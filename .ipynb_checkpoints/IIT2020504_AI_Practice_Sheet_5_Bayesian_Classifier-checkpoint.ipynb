```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The given Training Data Set

```python
data = pd.read_csv('training.csv')
data
```

From the given dataset, only Gender, Age, EstimatedSalary, Purchased coumns are useful.

```python
data = data[["Gender", "Age", "EstimatedSalary", "Purchased"]]
data
```

Ploting the histogram for the Age and EstimatedSalary column to check the distribution pattern.

We can observe that Age and EstimatedSalary follows normal disstribution (not acurately), so we can use normal distribution probability formula to directly calculate the probabilities.

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)
sns.histplot(data, ax=axes[0], x="Age", kde=True, color="r")
sns.histplot(data, ax=axes[1], x="EstimatedSalary", kde=True, color="b")
```

```python
y_zero_gender_male_prob = 0
y_one_gender_male_prob = 0
y_zero_gender_female_prob = 0
y_one_gender_female_prob = 0
y_one = 0
y_zero = 0

lst1 = list(data["Gender"])
lst2 = list(data["Purchased"])
sizee = len(lst1)
for i in range(sizee):
  if lst1[i] == "Male" and lst2[i] == 0:
    y_zero_gender_male_prob += 1
  if lst1[i] == "Female" and lst2[i] == 0:
    y_zero_gender_female_prob += 1
  if lst1[i] == "Male" and lst2[i] == 1:
    y_one_gender_male_prob += 1
  if lst1[i] == "Female" and lst2[i] == 1:
    y_one_gender_female_prob += 1

y_zero_gender_male_prob /= sizee
y_one_gender_male_prob /= sizee
y_zero_gender_female_prob /= sizee
y_one_gender_male_prob /= sizee

y_one = y_one_gender_male_prob + y_one_gender_female_prob
y_zero = y_zero_gender_male_prob + y_zero_gender_female_prob

print(y_zero_gender_male_prob)  # probability that gender is male and purchased is 0   
print(y_one_gender_male_prob)   # probability that gender is male and purchased is 1
print(y_zero_gender_female_prob) # probability that gender is female and purchased is 0
print(y_one_gender_male_prob)    # probability that gender is female and purchased is 1
```

```python

# function to calculate normal distribution probabilities
def calculate_likelihood_gaussian(df, feat_name, feat_val, Y, label):
  feat = list(df.columns)
  df = df[df[Y]==label]
  mean, std = df[feat_name].mean(), df[feat_name].std()
  p_x_given_y = (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-((feat_val-mean)**2 / (2 * std**2)))
  return p_x_given_y
```

Given Testing Data

```python
testing_data = pd.read_csv('testing.csv')
testing_data
```

Calulating probabilities using the fact that Age, EstimatedSalary follows normal distribution

```python

y_true = []
y_pred = []

for index, row in testing_data.iterrows():
  prob_y_one = y_one
  prob_y_zero = y_zero

  y_true.append(row['Purchased'])
  if row['Gender'] == "Female":
    prob_y_one *= y_one_gender_female_prob
    prob_y_zero *= y_zero_gender_female_prob
  else:
    prob_y_one *= y_one_gender_male_prob
    prob_y_zero *= y_zero_gender_male_prob
  
  prob_y_one *= calculate_likelihood_gaussian(data, "Age", row["Age"], "Purchased", 1)
  prob_y_zero *= calculate_likelihood_gaussian(data, "Age", row["Age"], "Purchased", 0)
  prob_y_one *= calculate_likelihood_gaussian(data, "EstimatedSalary", row["EstimatedSalary"], "Purchased", 1)
  prob_y_zero *= calculate_likelihood_gaussian(data, "EstimatedSalary", row["EstimatedSalary"], "Purchased", 0)

  # print(prob_y_one, prob_y_zero)
  if(prob_y_one >= prob_y_zero):
    # print(1)
    y_pred.append(1)
  else:
    # print(0)
    y_pred.append(0)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
print(f'Accuracy Score is {accuracy_score(y_true, y_pred)}')
print('The confusion matrix is given by: ')
print(confusion_matrix(y_true, y_pred))
print(f'F1 Score is {f1_score(y_true, y_pred)}')
```

Calculating probabilities using Test Categorical Model (without considering that Age, EstimatedSalary follows Normal distribution)

```python
def prob_age_given_y(df, age_val, age_err, label):
  lst = list(df['Age'])
  Y = list(df['Purchased'])
  cnt_age = 0
  cnt_label = 0

  for i in range(len(lst)):
    if Y[i] == label:
      cnt_label += 1
      if abs(age_val-lst[i]) <= age_err:
        cnt_age += 1
  return cnt_age/cnt_label

def prob_salary_given_y(df, sal_val, sal_err, label):
  lst = list(df['EstimatedSalary'])
  Y = list(df['Purchased'])
  cnt_sal = 0
  cnt_label = 0

  for i in range(len(lst)):
    if Y[i] == label:
      cnt_label += 1
      if abs(sal_val-lst[i]) <= sal_err:
        cnt_sal += 1
  return cnt_sal/cnt_label
```

```python

y_true = []
y_pred = []

for index, row in testing_data.iterrows():
  prob_y_one = y_one
  prob_y_zero = y_zero

  y_true.append(row['Purchased'])
  if row['Gender'] == "Female":
    prob_y_one *= y_one_gender_female_prob
    prob_y_zero *= y_zero_gender_female_prob
  else:
    prob_y_one *= y_one_gender_male_prob
    prob_y_zero *= y_zero_gender_male_prob
  
  prob_y_one *= prob_age_given_y(data, row['Age'], 3, 1)
  prob_y_zero *= prob_age_given_y(data, row['Age'], 3, 0)
  prob_y_one *= prob_salary_given_y(data, row['EstimatedSalary'], 4000, 1)
  prob_y_zero *= prob_salary_given_y(data, row['EstimatedSalary'], 4000, 0)

  # print(prob_y_one, prob_y_zero)
  if(prob_y_one >= prob_y_zero):
    # print(1)
    y_pred.append(1)
  else:
    # print(0)
    y_pred.append(0)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
print(f'Accuracy Score is {accuracy_score(y_true, y_pred)}')
print('The confusion matrix is given by: ')
print(confusion_matrix(y_true, y_pred))
print(f'F1 Score is {f1_score(y_true, y_pred)}')
```
